### Data Preprocessing

#### Split Dataset

```python
# Target directory
base_dir = '/content/garbage_dataset_split'
os.makedirs(base_dir, exist_ok=True)

splits = ['train', 'val', 'test']
```

```python
# Buat folder struktur split
for split in splits:
    for cls in class_names:
        os.makedirs(os.path.join(base_dir, split, cls), exist_ok=True)
```

```python
# Visualisasi distribusi kelas
def visualize_class_distribution(dataset_path, class_names):
    class_counts = []
    for cls in class_names:
        cls_path = os.path.join(dataset_path, cls)
        if os.path.exists(cls_path):
            count = len([f for f in os.listdir(cls_path) if f.endswith(('.jpg', '.jpeg', '.png'))])
            class_counts.append(count)
        else:
            class_counts.append(0)
            print(f"Warning: Class directory {cls_path} not found!")

    plt.figure(figsize=(12, 6))
    plt.bar(class_names, class_counts)
    plt.title('Distribution of Images per Class')
    plt.xlabel('Class Label')
    plt.ylabel('Number of Images')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Return data for further use
    return dict(zip(class_names, class_counts))
```

```python
# Visualisasi distribusi
class_distribution = visualize_class_distribution(dataset_path, class_names)
print("Class distribution:", class_distribution)
```

```python
# Fungsi split per kelas dengan stratified sampling
def split_class_images(cls_name, source_dir, dest_base, train_size=0.7, val_size=0.15):
    cls_path = os.path.join(source_dir, cls_name)
    if not os.path.exists(cls_path):
        print(f"Skipping {cls_name}, directory not found!")
        return

    images = [f for f in os.listdir(cls_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

    # Stratified split untuk memastikan distribusi seimbang
    train_imgs, temp_imgs = train_test_split(images, train_size=train_size, random_state=42)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

    print(f"Kelas {cls_name}: {len(images)} gambar ‚Üí Train: {len(train_imgs)}, Val: {len(val_imgs)}, Test: {len(test_imgs)}")

    # Copy gambar ke folder tujuan
    for img in train_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(dest_base, 'train', cls_name, img))
    for img in val_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(dest_base, 'val', cls_name, img))
    for img in test_imgs:
        shutil.copy(os.path.join(cls_path, img), os.path.join(dest_base, 'test', cls_name, img))
```

```python
# Lakukan split semua kelas
for cls in class_names:
    split_class_images(cls, dataset_path, base_dir)
```

#### Augmentasi Gambar

```python
# Hitung class weights untuk mengatasi imbalance
def calculate_class_weights(directory):
    class_counts = {}
    for cls in class_names:
        cls_dir = os.path.join(directory, cls)
        if os.path.exists(cls_dir):
            class_counts[cls] = len([f for f in os.listdir(cls_dir) if f.endswith(('.jpg', '.jpeg', '.png'))])
        else:
            class_counts[cls] = 0

    total = sum(class_counts.values())
    # Inverse weighting - kelas dengan lebih sedikit sampel mendapat bobot lebih tinggi
    weights = {i: total / (len(class_counts) * count) if count > 0 else 1.0
               for i, (cls, count) in enumerate(class_counts.items())}
    return weights
```

```python
# Augmentasi data yang lebih agresif untuk kelas minoritas
train_datagen_minority = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    zoom_range=0.3,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.7, 1.3],
    fill_mode='nearest'
)

# Augmentasi data standar untuk kelas mayoritas
train_datagen_majority = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Validasi dan test tanpa augmentasi
val_test_datagen = ImageDataGenerator(rescale=1./255)

img_size = (224, 224)
batch_size = 32  # Lebih kecil untuk Kaggle
```

```python
# Fungsi untuk mengidentifikasi kelas mayoritas (textile) vs minoritas
def is_minority_class(class_name):
    return class_name != 'textile'  # textile adalah kelas mayoritas

# Memuat data dengan augmentasi yang disesuaikan per kelas
def create_balanced_generator(directory, batch_size):
    generators = []

    # Buat generator terpisah untuk setiap kelas
    for cls in class_names:
        class_dir = os.path.join(directory, cls)
        if is_minority_class(cls):
            gen = train_datagen_minority.flow_from_directory(
                directory, # Use the parent directory
                target_size=img_size,
                batch_size=batch_size // len(class_names),
                class_mode='categorical', # Ensure class_mode is 'categorical' for one-hot encoding
                # Removed the classes argument as it was interfering with the label generation
                shuffle=True
            )
        else:
            gen = train_datagen_majority.flow_from_directory(
                directory, # Use the parent directory
                target_size=img_size,
                batch_size=batch_size // len(class_names),
                class_mode='categorical', # Ensure class_mode is 'categorical' for one-hot encoding
                # Removed the classes argument as it was interfering with the label generation
                shuffle=True
            )
        generators.append(gen)

    # Menggabungkan batch dari semua generator
    while True:
        batches = []
        for gen in generators:
            batch = next(gen)
            batches.append(batch)

        # Menggabungkan batch dari semua kelas
        x = np.concatenate([batch[0] for batch in batches])
        y = np.concatenate([batch[1] for batch in batches])

        # Shuffle batch gabungan
        indices = np.random.permutation(len(x))
        yield x[indices], y[indices]
```

```python
# Buat generator untuk training yang balanced
train_generator = create_balanced_generator(os.path.join(base_dir, 'train'), batch_size)
```

```python
# Generator standard untuk validasi dan testing
val_generator = val_test_datagen.flow_from_directory(
    os.path.join(base_dir, 'val'),
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = val_test_datagen.flow_from_directory(
    os.path.join(base_dir, 'test'),
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # untuk evaluasi model
)
```

## Modelling

```python
def create_model(input_shape=(224, 224, 3), num_classes=8):
    model = Sequential([
        # Block 1
        Conv2D(64, (3, 3), padding='same', activation='relu', input_shape=input_shape),
        Conv2D(64, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),

        # Block 2
        Conv2D(128, (3, 3), padding='same', activation='relu'),
        Conv2D(128, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.2),

        # Block 3
        Conv2D(256, (3, 3), padding='same', activation='relu'),
        Conv2D(256, (3, 3), padding='same', activation='relu'),
        Conv2D(256, (3, 3), padding='same', activation='relu'),  # Tambah satu layer konvolusi lagi
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.3),

        # Block 4
        Conv2D(512, (3, 3), padding='same', activation='relu'),
        Conv2D(512, (3, 3), padding='same', activation='relu'),
        Conv2D(512, (3, 3), padding='same', activation='relu'),  # Tambah satu layer konvolusi lagi
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.4),

        # Block 5
        Conv2D(512, (3, 3), padding='same', activation='relu'),
        Conv2D(512, (3, 3), padding='same', activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.4),

        # Fully connected layers
        Flatten(),
        Dense(1024, activation='relu'),  # Lebih banyak neuron
        BatchNormalization(),
        Dropout(0.5),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    return model
```

```python
# Buat model
model = create_model(input_shape=(img_size[0], img_size[1], 3), num_classes=len(class_names))
```

```python
# Compile model
optimizer = Adam(learning_rate=0.0001)
model.compile(
    loss='categorical_crossentropy',
    optimizer=optimizer,
    metrics=['accuracy']
)
```

```python
# Tampilkan ringkasan model
model.summary()
```

```python
# Callback untuk menyimpan model terbaik dan mencegah overfitting
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=15,  # Lebih banyak patience untuk training yang lebih lama
        restore_best_weights=True,
        verbose=1
    ),
    ModelCheckpoint(
        filepath='/kaggle/working/best_model.keras',  # lokasi writable
        monitor='val_loss',
        save_best_only=True,
        mode='min',
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-6,
        verbose=1
    )
]
```

```python
# Hitung jumlah step per epoch
steps_per_epoch = sum([len(os.listdir(os.path.join(base_dir, 'train', cls)))
                      for cls in class_names if os.path.exists(os.path.join(base_dir, 'train', cls))]) // batch_size
```

## Evaluasi dan Visualisasi

```python
# Evaluasi model
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test accuracy: {test_acc*100:.2f}%")
print(f"Test loss: {test_loss:.4f}")
```

```python
# Plot history training
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Evaluasi per kelas
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes
```

```python
# Confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
